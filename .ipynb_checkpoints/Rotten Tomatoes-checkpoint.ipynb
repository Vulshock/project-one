```python
#Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
# Study data files
```

```python
#Import CSV's
movies = ("rotten_tomatoes_movies.csv")
tmdb = ("tmdb_movies_data.csv")
#Convert to DataFrames
movie_data = pd.read_csv(movies)
tmdb_data = pd.read_csv(tmdb)
#Preview Rotten Tomatoes data
movie_data.head()
```

```python
#Preview IMDB data
tmdb_data.head()
```

```python
#create new column
tmdb_data["Title Release"] = ""
#change runtime to string
tmdb_data["runtime"] = tmdb_data["runtime"].astype(str)
#add values for new column
tmdb_data["Title Release"] = tmdb_data["Movie Title"] + " " + tmdb_data["runtime"] +".0"
tmdb_data.head()
```

```python
#Rename DataFrame column names
movie_data = movie_data.rename(columns={"rotten_tomatoes_link": "Link",
                                       "movie_info": "Movie Info", "studio_name": "Studio", "tomatometer_status": "Tomatometer Status",
                                       "tomatometer_rating": "Tomatometer Rating", "runtime_in_minutes": "Runtime", "in_theaters_date": "Theater Release", "audience_status": "Audience Status",
                                       "audience_rating": "Audience Rating", "audience_count": "Audience Count", "audience_top_critics_count": "Audience Top Critics Count",
                                       "audience_fresh_critics_count": " Audience Fresh Critics Count", "audience_rotten_critics_count": "Audience Rotten Critics Count", "tomatometer_count": "Tomatometer Count"})
movie_data.head(3)
```

```python
#create new column
movie_data["Title Release"] = ""

#change runtime to string
movie_data["Runtime"] = movie_data["Runtime"].astype(str)

#add values for new column
movie_data["Title Release"] = movie_data["Movie Title"] + " " + movie_data["Runtime"]
movie_data.head()
```

```python
#Merge data
combined_movie = pd.merge(movie_data, tmdb_data, how ='left', on='Title Release')

combined_movie
```

```python
#Rename Budge and Revenue columns
combined_movie = combined_movie.rename(columns={"budget": "Budget", "revenue": "Revenue"})
#Drop all column with NaN in Budget and Revenue
combined_movie = combined_movie.dropna(subset=['Budget', "Revenue"])
combined_movie
```

```python
#Split columns values into new columns
combined_movie['Genre_One'] = ""
combined_movie["Year"] = ""
combined_movie["Year"] = combined_movie["release_date"].apply(lambda x:x.split('/')[2])
combined_movie['Genre_One'] = combined_movie["genres"].apply(lambda x: x.split('|')[0])
# combined_movie[combined_movie["genres"].str.contains("Comedy")]
combined_movie = combined_movie.rename(columns={"Movie Title_x": "Movie Title"})
combined_movie.columns
```

```python
comedy_df = combined_movie.loc[combined_movie["Genre_One"] == "Comedy", :]
comedy_df

grouped_years_comedy = comedy_df.groupby(["Year"]).mean()
#grouped_years_comedy
grouped_years_comedy_sum = comedy_df.groupby(["Year"]).sum()
grouped_years_comedy_sum
grouped_years_comedy["Total Revenue"] = grouped_years_comedy_sum["Revenue"]
grouped_years_comedy

cleaned_years_group_comedy = grouped_years_comedy.drop(["1972", "1962", "1963", "1965", "1968", "1969", "1971"])
cleaned_years_group_comedy
```

```python
x = cleaned_years_group_comedy["Tomatometer Rating"]
y = cleaned_years_group_comedy["Audience Rating"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x,y)
regression_value = x * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))


size = cleaned_years_group_comedy["Total Revenue"]/10000000
plt.title("Tomatometer Rating vs. Audience Rating of All Comedies in a Given Year")
plt.xlabel("Tomatometer Rating")
plt.ylabel("Audience Rating")
plt.plot(x, regression_value,"r-")
plt.annotate(line_eq, (70,50), fontsize=15, color="r")


plt.scatter(x, y, s=size, alpha=0.75)
plt.show()

plt.savefig("Comedy Ratings by Year Scatterplot")
```

```python
#New DataFrame
refined_movie_data = combined_movie[["Movie Title", "Studio","Runtime", "Theater Release","Budget", "Revenue", "Tomatometer Status", "Tomatometer Rating", "Tomatometer Count", "Audience Status", "Audience Rating", "Audience Count"]]
refined_movie_data
```

```python
#New DataFrame for only Disney Movies
Disney = refined_movie_data[(refined_movie_data["Studio"]== "Buena Vista") | \
                           (refined_movie_data["Studio"]== "Disney") |\
                           (refined_movie_data["Studio"]== "Walt Disney Pictures")]

Disney
```

```python
#Plot Scatter Plot
x_axis = Disney["Tomatometer Rating"]
y_axis = Disney["Audience Rating"]
plt.scatter( x_axis, y_axis)
```

```python
#Create linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)

regress_value = x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

#Plot new scatter plot
plt.scatter(x_axis, y_axis,marker='o')
plt.plot(x_axis, regress_value, "r-")
plt.annotate(line_eq,(60, 30), fontsize=15, color="red")

#Set x & y limiters
plt.xlim(0,110)
plt.ylim(0,100)

#Add labels
plt.title("Critic vs Audience Rating")
plt.xlabel("Critic Rating")
plt.ylabel("Audience Rating")

plt.savefig("Image folder/ Critic vs Audience Rating.png")
```

```python
#Create Box plots and set y limiter
Disney.boxplot('Audience Rating', by='Studio', figsize=(10,5))

#limit rating score 0 - 100
plt.ylim(0,100)

#label y-axis
plt.ylabel ("Audience Rating")
plt.savefig("Image folder/ Audience Rating per Studio.png")

Disney.boxplot('Tomatometer Rating', by='Studio', figsize=(10,5))

#limit rating score 0 - 100
plt.ylim(0,110)

#label y-axis
plt.ylabel("Critic Rating")

#save fig
plt.savefig("Image folder/ Critic per Studio.png")
```
